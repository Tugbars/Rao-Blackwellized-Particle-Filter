# RBPF-KSC Visualization

Visualize filter performance:
- Log-volatility tracking
- Regime detection accuracy
- **Outlier robustness** (Phase 6 vs Phase 7 critical test)
- Overshooting vs undershooting analysis
- Tail tracking

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams['figure.figsize'] = (16, 10)
plt.rcParams['font.size'] = 10
```

```python
# Load data
df = pd.read_csv('rbpf_visualization.csv')
print(f"Loaded {len(df)} observations")
print(f"Columns: {list(df.columns)}")
df.head()
```

## 1. Main Dashboard

```python
fig, axes = plt.subplots(5, 1, figsize=(16, 14), sharex=True)

# Panel 1: Log-Volatility
ax = axes[0]
ax.plot(df['t'], df['true_log_vol'], 'b-', label='True', linewidth=1.2)
ax.plot(df['t'], df['est_log_vol_fast'], 'r-', label='Fast (t)', linewidth=0.8, alpha=0.8)
ax.plot(df['t'], df['est_log_vol_smooth'], 'g-', label='Smooth (t-5)', linewidth=0.8, alpha=0.8)
ax.set_ylabel('Log-Volatility')
ax.legend(loc='upper right')
ax.set_title('RBPF-KSC: Full Timeseries', fontweight='bold', fontsize=14)

# Panel 2: Regime
ax = axes[1]
ax.step(df['t'], df['true_regime'], 'b-', where='post', label='True', linewidth=2)
ax.step(df['t'], df['est_regime_fast'], 'r--', where='post', label='Fast', linewidth=1, alpha=0.7)
ax.step(df['t'], df['est_regime_smooth'], 'g-', where='post', label='Smooth', linewidth=1.5, alpha=0.8)
ax.set_ylabel('Regime')
ax.set_yticks([0, 1, 2, 3])
ax.set_yticklabels(['Calm', 'Medium', 'High', 'Crisis'])
ax.legend(loc='upper right')

# Panel 3: Confidence
ax = axes[2]
ax.plot(df['t'], df['regime_confidence'], 'purple', linewidth=0.8)
ax.axhline(0.7, color='red', linestyle='--', alpha=0.5, label='70%')
ax.set_ylabel('Confidence')
ax.set_ylim(0, 1)
ax.legend(loc='upper right')

# Panel 4: Surprise
ax = axes[3]
ax.plot(df['t'], df['surprise'], 'orange', linewidth=0.6)
ax.axhline(5.0, color='red', linestyle='--', alpha=0.5, label='Alert')
ax.set_ylabel('Surprise')
ax.legend(loc='upper right')

# Panel 5: ESS
ax = axes[4]
ax.plot(df['t'], df['ess'], 'teal', linewidth=0.6)
ax.axhline(100, color='red', linestyle='--', alpha=0.5, label='Resample')
ax.set_ylabel('ESS')
ax.set_xlabel('Time')
ax.legend(loc='upper right')

plt.tight_layout()
plt.savefig('rbpf_main_dashboard.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 2. CRITICAL TEST: Outlier Burst vs Real Crash

This is the key test: Can the filter distinguish between:
- **Phase 6 (t=3500-3600)**: 20% outlier burst with NO real vol change  should NOT adapt
- **Phase 7 (t=3600-4000)**: Real crash to crisis regime  should adapt quickly

A good filter stays stable during Phase 6 but jumps in Phase 7.

```python
# Zoom into the critical region: t=3300 to t=4200
zoom_start, zoom_end = 3300, 4200
phase6_start, phase6_end = 3500, 3600  # Outlier burst (no real change)
phase7_start, phase7_end = 3600, 4000  # Real crash

zoom = df[(df['t'] >= zoom_start) & (df['t'] <= zoom_end)].copy()

fig, axes = plt.subplots(4, 1, figsize=(16, 12), sharex=True)

# Panel 1: Log-Volatility (zoomed)
ax = axes[0]
ax.axvspan(phase6_start, phase6_end, alpha=0.3, color='yellow', label='Phase 6: Outlier Burst')
ax.axvspan(phase7_start, phase7_end, alpha=0.3, color='red', label='Phase 7: Real Crash')
ax.plot(zoom['t'], zoom['true_log_vol'], 'b-', label='True', linewidth=2)
ax.plot(zoom['t'], zoom['est_log_vol_fast'], 'r-', label='Est (Fast)', linewidth=1.5, alpha=0.9)
ax.set_ylabel('Log-Volatility')
ax.legend(loc='upper left')
ax.set_title('CRITICAL TEST: Outlier Burst (Phase 6) vs Real Crash (Phase 7)', fontweight='bold', fontsize=14)

# Panel 2: Regime
ax = axes[1]
ax.axvspan(phase6_start, phase6_end, alpha=0.3, color='yellow')
ax.axvspan(phase7_start, phase7_end, alpha=0.3, color='red')
ax.step(zoom['t'], zoom['true_regime'], 'b-', where='post', label='True', linewidth=2)
ax.step(zoom['t'], zoom['est_regime_fast'], 'r--', where='post', label='Est', linewidth=1.5)
ax.set_ylabel('Regime')
ax.set_yticks([0, 1, 2, 3])
ax.set_yticklabels(['Calm', 'Med', 'High', 'Crisis'])
ax.legend(loc='upper left')

# Panel 3: Surprise (detect outliers vs real changes)
ax = axes[2]
ax.axvspan(phase6_start, phase6_end, alpha=0.3, color='yellow')
ax.axvspan(phase7_start, phase7_end, alpha=0.3, color='red')
ax.plot(zoom['t'], zoom['surprise'], 'orange', linewidth=1)
ax.axhline(5.0, color='black', linestyle='--', alpha=0.5, label='Alert threshold')
ax.set_ylabel('Surprise')
ax.legend(loc='upper right')

# Panel 4: Tracking error
ax = axes[3]
ax.axvspan(phase6_start, phase6_end, alpha=0.3, color='yellow')
ax.axvspan(phase7_start, phase7_end, alpha=0.3, color='red')
error = zoom['est_log_vol_fast'] - zoom['true_log_vol']
ax.fill_between(zoom['t'], 0, error, where=(error > 0), color='salmon', alpha=0.6)
ax.fill_between(zoom['t'], 0, error, where=(error < 0), color='lightblue', alpha=0.6)
ax.plot(zoom['t'], error, 'k-', linewidth=0.8)
ax.axhline(0, color='black', linewidth=1)
ax.set_ylabel('Error (est - true)')
ax.set_xlabel('Time')

plt.tight_layout()
plt.savefig('rbpf_outlier_test.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# Quantify Phase 6 vs Phase 7 performance
phase6 = df[(df['t'] >= phase6_start) & (df['t'] < phase6_end)]
phase7 = df[(df['t'] >= phase7_start) & (df['t'] < phase7_end)]
pre_burst = df[(df['t'] >= 3400) & (df['t'] < phase6_start)]  # Baseline before burst

print("=" * 60)
print("OUTLIER ROBUSTNESS TEST RESULTS")
print("=" * 60)

# Phase 6: Should NOT move estimate
p6_est_start = phase6['est_log_vol_fast'].iloc[0]
p6_est_end = phase6['est_log_vol_fast'].iloc[-1]
p6_est_max = phase6['est_log_vol_fast'].max()
p6_true = phase6['true_log_vol'].iloc[0]  # Should be constant at -4.0
p6_movement = p6_est_max - p6_est_start

print(f"\nPhase 6 (Outlier Burst, t={phase6_start}-{phase6_end}):")
print(f"  True log-vol:        {p6_true:.3f} (constant, no real change)")
print(f"  Est start:           {p6_est_start:.3f}")
print(f"  Est max:             {p6_est_max:.3f}")
print(f"  Est movement:        {p6_movement:+.3f}")
print(f"  Surprise spikes:     {(phase6['surprise'] > 5).sum()} / {len(phase6)}")

if p6_movement < 0.3:
    print(f"   PASS: Filter stayed stable (movement < 0.3)")
elif p6_movement < 0.5:
    print(f"   MARGINAL: Some movement but contained")
else:
    print(f"   FAIL: Filter chased outliers (movement >= 0.5)")

# Phase 7: Should adapt quickly
p7_true = phase7['true_log_vol'].iloc[0]  # Should be -2.0
p7_est_start = phase7['est_log_vol_fast'].iloc[0]
p7_est_50 = phase7['est_log_vol_fast'].iloc[min(50, len(phase7)-1)]  # 50 ticks in
p7_est_100 = phase7['est_log_vol_fast'].iloc[min(100, len(phase7)-1)]  # 100 ticks in
p7_target_gap = abs(p7_true - p7_est_start)  # How far we need to move
p7_closed_50 = abs(p7_true - p7_est_50)  # Gap after 50 ticks
p7_closed_100 = abs(p7_true - p7_est_100)  # Gap after 100 ticks

print(f"\nPhase 7 (Real Crash, t={phase7_start}-{phase7_end}):")
print(f"  True log-vol:        {p7_true:.3f} (crisis level)")
print(f"  Est at t={phase7_start}:       {p7_est_start:.3f}")
print(f"  Initial gap:         {p7_target_gap:.3f}")
print(f"  Gap after 50 ticks:  {p7_closed_50:.3f} ({100*(1-p7_closed_50/p7_target_gap):.0f}% closed)")
print(f"  Gap after 100 ticks: {p7_closed_100:.3f} ({100*(1-p7_closed_100/p7_target_gap):.0f}% closed)")

if p7_closed_100 < 0.3:
    print(f"   PASS: Filter adapted within 100 ticks")
elif p7_closed_100 < 0.5:
    print(f"   MARGINAL: Slow but adapting")
else:
    print(f"   FAIL: Filter too slow to adapt")

print("\n" + "=" * 60)
```

## 3. Volatility Tracking

```python
fig, axes = plt.subplots(2, 1, figsize=(16, 8), sharex=True)

# Panel 1: Volatility in linear scale
ax = axes[0]
ax.plot(df['t'], df['true_vol'], 'b-', label='True Vol', linewidth=1.2)
ax.plot(df['t'], df['est_vol_fast'], 'r-', label='Est Vol (Fast)', linewidth=0.8, alpha=0.8)
ax.set_ylabel('Volatility')
ax.legend(loc='upper right')
ax.set_title('Volatility Tracking (Linear Scale)', fontweight='bold', fontsize=14)

# Panel 2: Error
ax = axes[1]
error = df['est_log_vol_fast'] - df['true_log_vol']
ax.fill_between(df['t'], 0, error, where=(error > 0), color='salmon', alpha=0.5, label='Overestimate')
ax.fill_between(df['t'], 0, error, where=(error < 0), color='lightblue', alpha=0.5, label='Underestimate')
ax.plot(df['t'], error, 'k-', linewidth=0.5)
ax.axhline(0, color='black', linewidth=0.5)
ax.set_ylabel('Log-Vol Error')
ax.set_xlabel('Time')
ax.legend(loc='upper right')

plt.tight_layout()
plt.savefig('rbpf_vol_tracking.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 4. Regime Detection Analysis

```python
try:
    from sklearn.metrics import confusion_matrix
    import seaborn as sns
    
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))
    
    cm_fast = confusion_matrix(df['true_regime'], df['est_regime_fast'], labels=[0, 1, 2, 3])
    cm_fast_norm = cm_fast.astype('float') / cm_fast.sum(axis=1)[:, np.newaxis]
    cm_fast_norm = np.nan_to_num(cm_fast_norm)
    
    sns.heatmap(cm_fast_norm, annot=True, fmt='.2f', cmap='Blues', ax=axes[0],
                xticklabels=['Calm', 'Med', 'High', 'Crisis'],
                yticklabels=['Calm', 'Med', 'High', 'Crisis'])
    axes[0].set_title('Regime Confusion (Fast Signal)')
    axes[0].set_xlabel('Estimated')
    axes[0].set_ylabel('True')
    
    cm_smooth = confusion_matrix(df['true_regime'], df['est_regime_smooth'], labels=[0, 1, 2, 3])
    cm_smooth_norm = cm_smooth.astype('float') / cm_smooth.sum(axis=1)[:, np.newaxis]
    cm_smooth_norm = np.nan_to_num(cm_smooth_norm)
    
    sns.heatmap(cm_smooth_norm, annot=True, fmt='.2f', cmap='Greens', ax=axes[1],
                xticklabels=['Calm', 'Med', 'High', 'Crisis'],
                yticklabels=['Calm', 'Med', 'High', 'Crisis'])
    axes[1].set_title('Regime Confusion (Smooth Signal)')
    axes[1].set_xlabel('Estimated')
    axes[1].set_ylabel('True')
    
    plt.tight_layout()
    plt.savefig('rbpf_regime_confusion.png', dpi=150, bbox_inches='tight')
    plt.show()
except ImportError:
    print("sklearn/seaborn not available, skipping confusion matrix")

print(f"Fast Signal Accuracy:   {(df['true_regime'] == df['est_regime_fast']).mean() * 100:.1f}%")
print(f"Smooth Signal Accuracy: {(df['true_regime'] == df['est_regime_smooth']).mean() * 100:.1f}%")
```

## 5. Overshooting vs Undershooting

```python
df['log_vol_error'] = df['est_log_vol_fast'] - df['true_log_vol']
df['abs_log_vol_error'] = df['log_vol_error'].abs()
df['true_vol_diff'] = df['true_log_vol'].diff()
df['vol_rising'] = df['true_vol_diff'] > 0.01
df['vol_falling'] = df['true_vol_diff'] < -0.01

rising = df[df['vol_rising']]
falling = df[df['vol_falling']]

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Error when vol rising
ax = axes[0]
ax.hist(rising['log_vol_error'], bins=40, alpha=0.7, color='red')
ax.axvline(0, color='black', linestyle='--')
ax.axvline(rising['log_vol_error'].mean(), color='red', linestyle='-', linewidth=2,
           label=f'Mean: {rising["log_vol_error"].mean():.3f}')
ax.set_xlabel('Error (est - true)')
ax.set_title(f'Error When Vol RISING (n={len(rising)})')
ax.legend()
label = 'UNDERSHOOTING' if rising['log_vol_error'].mean() < 0 else 'OVERSHOOTING'
ax.text(0.95, 0.95, label, transform=ax.transAxes, ha='right', va='top',
        fontsize=12, fontweight='bold', color='blue' if 'UNDER' in label else 'red')

# Error when vol falling
ax = axes[1]
ax.hist(falling['log_vol_error'], bins=40, alpha=0.7, color='green')
ax.axvline(0, color='black', linestyle='--')
ax.axvline(falling['log_vol_error'].mean(), color='green', linestyle='-', linewidth=2,
           label=f'Mean: {falling["log_vol_error"].mean():.3f}')
ax.set_xlabel('Error (est - true)')
ax.set_title(f'Error When Vol FALLING (n={len(falling)})')
ax.legend()
label = 'OVERSHOOTING' if falling['log_vol_error'].mean() > 0 else 'UNDERSHOOTING'
ax.text(0.95, 0.95, label, transform=ax.transAxes, ha='right', va='top',
        fontsize=12, fontweight='bold', color='red' if 'OVER' in label else 'blue')

plt.tight_layout()
plt.savefig('rbpf_overshoot.png', dpi=150, bbox_inches='tight')
plt.show()

print(f"Rising vol bias:  {rising['log_vol_error'].mean():+.4f}")
print(f"Falling vol bias: {falling['log_vol_error'].mean():+.4f}")
```

## 6. Tail Tracking

```python
vol_p90 = df['true_vol'].quantile(0.90)
df['is_tail'] = df['true_vol'] >= vol_p90
normal = df[~df['is_tail']]
tail = df[df['is_tail']]

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Scatter: true vs estimated
ax = axes[0]
ax.scatter(normal['true_vol'], normal['est_vol_fast'], alpha=0.3, s=5, label='Normal', color='blue')
ax.scatter(tail['true_vol'], tail['est_vol_fast'], alpha=0.5, s=15, label='Tail (top 10%)', color='red')
max_vol = max(df['true_vol'].max(), df['est_vol_fast'].max())
ax.plot([0, max_vol], [0, max_vol], 'k--', linewidth=1, label='Perfect')
ax.set_xlabel('True Volatility')
ax.set_ylabel('Estimated Volatility')
ax.set_title('True vs Estimated Volatility')
ax.legend()

# MAE by regime
ax = axes[1]
regime_labels = ['Calm', 'Medium', 'High', 'Crisis']
mae_by_regime = [df[df['true_regime'] == r]['abs_log_vol_error'].mean() for r in range(4)]
colors = ['lightgreen', 'lightblue', 'orange', 'salmon']
bars = ax.bar(regime_labels, mae_by_regime, color=colors)
ax.set_ylabel('Mean Absolute Error')
ax.set_title('MAE by True Regime')
for bar, val in zip(bars, mae_by_regime):
    if not np.isnan(val):
        ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01, 
                f'{val:.3f}', ha='center', va='bottom')

plt.tight_layout()
plt.savefig('rbpf_tail_tracking.png', dpi=150, bbox_inches='tight')
plt.show()

print(f"Normal MAE: {normal['abs_log_vol_error'].mean():.4f}")
print(f"Tail MAE:   {tail['abs_log_vol_error'].mean():.4f}")
print(f"Tail Corr:  {tail['true_vol'].corr(tail['est_vol_fast']):.4f}")
```

## 7. Summary & Diagnosis

```python
print("=" * 60)
print("RBPF-KSC PERFORMANCE SUMMARY")
print("=" * 60)

vol_corr = df['true_vol'].corr(df['est_vol_fast'])
log_vol_mae = df['abs_log_vol_error'].mean()
log_vol_rmse = np.sqrt((df['log_vol_error'] ** 2).mean())
fast_acc = (df['true_regime'] == df['est_regime_fast']).mean() * 100
smooth_acc = (df['true_regime'] == df['est_regime_smooth']).mean() * 100

print(f"\nVolatility Tracking:")
print(f"  Correlation:     {vol_corr:.4f}")
print(f"  Log-Vol MAE:     {log_vol_mae:.4f}")
print(f"  Log-Vol RMSE:    {log_vol_rmse:.4f}")
print(f"  Bias:            {df['log_vol_error'].mean():+.4f}")

print(f"\nRegime Detection:")
print(f"  Fast accuracy:   {fast_acc:.1f}%")
print(f"  Smooth accuracy: {smooth_acc:.1f}%")

print(f"\nESS Health:")
print(f"  Mean ESS:        {df['ess'].mean():.1f}")
print(f"  Min ESS:         {df['ess'].min():.1f}")

print(f"\nTail Tracking (top 10%):")
print(f"  Tail MAE:        {tail['abs_log_vol_error'].mean():.4f}")
print(f"  Tail Corr:       {tail['true_vol'].corr(tail['est_vol_fast']):.4f}")

print("\n" + "-" * 60)
print("DIAGNOSIS:")
print("-" * 60)

if fast_acc < 50:
    print(f" LOW regime accuracy ({fast_acc:.1f}%)")
else:
    print(f" Regime accuracy OK ({fast_acc:.1f}%)")

if log_vol_mae > 0.5:
    print(f" HIGH log-vol error (MAE={log_vol_mae:.3f})")
else:
    print(f" Log-vol tracking OK (MAE={log_vol_mae:.3f})")

rising_bias = rising['log_vol_error'].mean()
if rising_bias < -0.1:
    print(f" UNDERSHOOTING during vol spikes (bias={rising_bias:.3f})")
else:
    print(f" Vol spike tracking OK (bias={rising_bias:.3f})")

if df['ess'].min() < 30:
    print(f" ESS drops very low (min={df['ess'].min():.1f})")
else:
    print(f" ESS healthy (min={df['ess'].min():.1f})")

print("\n" + "=" * 60)
```
