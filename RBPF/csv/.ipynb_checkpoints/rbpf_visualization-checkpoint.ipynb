# RBPF Performance Visualization

This notebook visualizes how well the Rao-Blackwellized Particle Filter:
1. **Tracks volatility** (h_est vs h_true)
2. **Detects regime changes** (regime_est vs regime_true)
3. **Handles outliers** (does NOT overreact to noise)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.collections import LineCollection
import warnings
warnings.filterwarnings('ignore')

# Style
plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams['figure.figsize'] = (14, 5)
plt.rcParams['font.size'] = 10
```

## Load Data

```python
# Load timeseries
df = pd.read_csv('timeseries.csv')
phases = pd.read_csv('phases.csv')

print(f"Loaded {len(df)} ticks across {df['phase_id'].nunique()} phases")
print(f"\nPhases:")
display(phases[['phase_id', 'name', 'start_tick', 'end_tick', 'description']])
```

## 1. Volatility Tracking Overview

Does the filter follow true volatility without overshooting or undershooting?

```python
fig, axes = plt.subplots(3, 1, figsize=(16, 12), sharex=True)

# --- Panel 1: Log-volatility tracking ---
ax = axes[0]
ax.plot(df['t'], df['h_true'], 'k-', label='True h', linewidth=1.2)
ax.plot(df['t'], df['h_est'], 'b-', label='Estimated h', linewidth=0.8, alpha=0.8)
ax.fill_between(df['t'], 
                df['h_est'] - 2*df['h_std'], 
                df['h_est'] + 2*df['h_std'], 
                alpha=0.2, color='blue', label='95% CI')

# Mark outliers
outliers = df[df['is_outlier'] == 1]
ax.scatter(outliers['t'], outliers['h_true'], c='red', s=15, alpha=0.6, 
           marker='x', label=f'Outliers ({len(outliers)})', zorder=5)

# Phase boundaries
colors = plt.cm.Set3(np.linspace(0, 1, len(phases)))
for i, phase in phases.iterrows():
    ax.axvline(phase['start_tick'], color='gray', linestyle='--', alpha=0.4, linewidth=0.8)
    ax.axvspan(phase['start_tick'], phase['end_tick'], alpha=0.1, color=colors[i])
    ax.text(phase['start_tick'] + 20, ax.get_ylim()[1] - 0.2, 
            phase['name'], fontsize=8, rotation=45, ha='left')

ax.set_ylabel('Log-volatility (h)')
ax.set_title('Volatility Tracking: True vs Estimated', fontweight='bold')
ax.legend(loc='lower right')

# --- Panel 2: Tracking error ---
ax = axes[1]
error = df['h_est'] - df['h_true']
ax.plot(df['t'], error, 'purple', linewidth=0.6, alpha=0.7)
ax.axhline(0, color='black', linestyle='-', linewidth=0.5)
ax.axhline(0.3, color='red', linestyle='--', linewidth=0.5, alpha=0.5)
ax.axhline(-0.3, color='red', linestyle='--', linewidth=0.5, alpha=0.5)

# Color by phase
for i, phase in phases.iterrows():
    ax.axvspan(phase['start_tick'], phase['end_tick'], alpha=0.1, color=colors[i])

ax.set_ylabel('Error (h_est - h_true)')
ax.set_title('Tracking Error Over Time', fontweight='bold')
ax.set_ylim(-1.5, 1.5)

# --- Panel 3: ESS and Outlier Fraction ---
ax = axes[2]
ax2 = ax.twinx()

ax.plot(df['t'], df['ess'], 'green', linewidth=0.6, alpha=0.7, label='ESS')
ax.axhline(50, color='red', linestyle='--', linewidth=0.5, alpha=0.5)
ax.set_ylabel('ESS', color='green')
ax.set_ylim(0, 260)

ax2.fill_between(df['t'], 0, df['outlier_frac'], alpha=0.3, color='orange', label='Outlier Frac')
ax2.set_ylabel('Outlier Fraction', color='orange')
ax2.set_ylim(0, 1)

ax.set_xlabel('Time (tick)')
ax.set_title('Filter Health: ESS & Outlier Detection', fontweight='bold')

plt.tight_layout()
plt.savefig('volatility_tracking.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 2. Regime Detection

How quickly and accurately does the filter detect regime changes?

```python
fig, axes = plt.subplots(2, 1, figsize=(16, 8), sharex=True)

# --- Panel 1: Regime comparison ---
ax = axes[0]
ax.step(df['t'], df['regime_true'], 'k-', where='post', label='True regime', linewidth=2)
ax.step(df['t'], df['regime_est'], 'b--', where='post', label='Estimated regime', linewidth=1.5, alpha=0.8)

# Mark disagreements
disagree = df[df['regime_true'] != df['regime_est']]
ax.scatter(disagree['t'], disagree['regime_true'], c='red', s=3, alpha=0.3, label='Mismatch')

ax.set_ylabel('Regime')
ax.set_yticks([0, 1, 2, 3])
ax.set_yticklabels(['R0 (Calm)', 'R1 (Normal)', 'R2 (Elevated)', 'R3 (Crisis)'])
ax.set_title('Regime Detection: True vs Estimated', fontweight='bold')
ax.legend(loc='upper right')

# --- Panel 2: Regime probabilities ---
ax = axes[1]
ax.stackplot(df['t'], 
             df['regime_prob_0'], df['regime_prob_1'], 
             df['regime_prob_2'], df['regime_prob_3'],
             labels=['R0 (Calm)', 'R1 (Normal)', 'R2 (Elevated)', 'R3 (Crisis)'],
             alpha=0.7)

ax.set_ylabel('Regime Probability')
ax.set_xlabel('Time (tick)')
ax.set_title('Regime Probability Distribution Over Time', fontweight='bold')
ax.legend(loc='upper right')
ax.set_ylim(0, 1)

plt.tight_layout()
plt.savefig('regime_detection.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 3. Phase-by-Phase Analysis

How does the filter perform in each test phase?

```python
# Compute metrics per phase
phase_metrics = []

for phase_id in df['phase_id'].unique():
    phase_df = df[df['phase_id'] == phase_id]
    phase_info = phases[phases['phase_id'] == phase_id].iloc[0]
    
    error = phase_df['h_est'] - phase_df['h_true']
    
    metrics = {
        'phase_id': phase_id,
        'name': phase_info['name'],
        'n_ticks': len(phase_df),
        'rmse': np.sqrt((error ** 2).mean()),
        'bias': error.mean(),
        'std': error.std(),
        'regime_accuracy': (phase_df['regime_true'] == phase_df['regime_est']).mean() * 100,
        'mean_ess': phase_df['ess'].mean(),
        'min_ess': phase_df['ess'].min(),
        'ess_collapses': (phase_df['ess'] < 50).sum(),
        'n_outliers': phase_df['is_outlier'].sum(),
        'mean_outlier_frac': phase_df['outlier_frac'].mean(),
        'coverage_95': ((phase_df['h_true'] >= phase_df['h_est'] - 2*phase_df['h_std']) & 
                        (phase_df['h_true'] <= phase_df['h_est'] + 2*phase_df['h_std'])).mean() * 100
    }
    phase_metrics.append(metrics)

metrics_df = pd.DataFrame(phase_metrics)
display(metrics_df.round(2))
```

```python
# Bar chart comparison
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# RMSE by phase
ax = axes[0, 0]
bars = ax.bar(metrics_df['name'], metrics_df['rmse'], color='steelblue', alpha=0.7)
ax.axhline(metrics_df['rmse'].mean(), color='red', linestyle='--', label=f"Mean: {metrics_df['rmse'].mean():.3f}")
ax.set_ylabel('RMSE')
ax.set_title('Tracking Error (RMSE) by Phase', fontweight='bold')
ax.tick_params(axis='x', rotation=45)
ax.legend()

# Regime accuracy by phase
ax = axes[0, 1]
colors = ['green' if x > 80 else 'orange' if x > 50 else 'red' for x in metrics_df['regime_accuracy']]
ax.bar(metrics_df['name'], metrics_df['regime_accuracy'], color=colors, alpha=0.7)
ax.axhline(80, color='green', linestyle='--', alpha=0.5)
ax.set_ylabel('Accuracy (%)')
ax.set_title('Regime Detection Accuracy by Phase', fontweight='bold')
ax.tick_params(axis='x', rotation=45)
ax.set_ylim(0, 105)

# Coverage by phase
ax = axes[1, 0]
colors = ['green' if 90 <= x <= 99 else 'orange' if 80 <= x <= 100 else 'red' for x in metrics_df['coverage_95']]
ax.bar(metrics_df['name'], metrics_df['coverage_95'], color=colors, alpha=0.7)
ax.axhline(95, color='black', linestyle='-', linewidth=2, label='Target: 95%')
ax.axhline(90, color='green', linestyle='--', alpha=0.5)
ax.axhline(100, color='green', linestyle='--', alpha=0.5)
ax.set_ylabel('Coverage (%)')
ax.set_title('95% CI Coverage by Phase (should be ~95%)', fontweight='bold')
ax.tick_params(axis='x', rotation=45)
ax.set_ylim(0, 105)
ax.legend()

# ESS by phase
ax = axes[1, 1]
ax.bar(metrics_df['name'], metrics_df['mean_ess'], color='purple', alpha=0.7, label='Mean ESS')
ax.scatter(metrics_df['name'], metrics_df['min_ess'], color='red', s=50, zorder=5, label='Min ESS')
ax.axhline(50, color='red', linestyle='--', alpha=0.5)
ax.set_ylabel('ESS')
ax.set_title('Effective Sample Size by Phase', fontweight='bold')
ax.tick_params(axis='x', rotation=45)
ax.legend()

plt.tight_layout()
plt.savefig('phase_analysis.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 4. Critical Test: Outlier Burst vs Real Change

**Key question**: Does the filter distinguish between outlier bursts (should NOT adapt) and real changes (should adapt)?

```python
# Compare Phase 6 (outlier_burst) vs Phase 7 (real_crash)
phase_6 = df[df['phase_name'] == 'outlier_burst']
phase_7 = df[df['phase_name'] == 'real_crash']

# Get some context before
t_start = phase_6['t'].min() - 100
t_end = phase_7['t'].max() + 100
window = df[(df['t'] >= t_start) & (df['t'] <= t_end)]

fig, axes = plt.subplots(3, 1, figsize=(14, 10), sharex=True)

# --- Panel 1: Log-vol ---
ax = axes[0]
ax.plot(window['t'], window['h_true'], 'k-', label='True h', linewidth=2)
ax.plot(window['t'], window['h_est'], 'b-', label='Estimated h', linewidth=1.5)
ax.fill_between(window['t'], 
                window['h_est'] - 2*window['h_std'], 
                window['h_est'] + 2*window['h_std'], 
                alpha=0.2, color='blue')

# Mark outliers
outliers = window[window['is_outlier'] == 1]
ax.scatter(outliers['t'], outliers['h_true'], c='red', s=30, marker='x', zorder=5)

# Phase regions
ax.axvspan(phase_6['t'].min(), phase_6['t'].max(), alpha=0.2, color='orange', label='Outlier Burst (NO change)')
ax.axvspan(phase_7['t'].min(), phase_7['t'].max(), alpha=0.2, color='red', label='Real Crash')

ax.set_ylabel('Log-volatility')
ax.set_title('CRITICAL TEST: Outlier Burst vs Real Change', fontweight='bold', fontsize=14)
ax.legend(loc='upper left')

# --- Panel 2: Tracking error ---
ax = axes[1]
error = window['h_est'] - window['h_true']
ax.plot(window['t'], error, 'purple', linewidth=1)
ax.axhline(0, color='black', linewidth=0.5)
ax.axvspan(phase_6['t'].min(), phase_6['t'].max(), alpha=0.2, color='orange')
ax.axvspan(phase_7['t'].min(), phase_7['t'].max(), alpha=0.2, color='red')

ax.set_ylabel('Error (h_est - h_true)')

# Annotate
error_phase6 = (phase_6['h_est'] - phase_6['h_true']).abs().mean()
error_phase7 = (phase_7['h_est'] - phase_7['h_true']).abs().mean()
ax.text(phase_6['t'].mean(), ax.get_ylim()[1] * 0.8, f'MAE: {error_phase6:.3f}', ha='center', fontsize=10, fontweight='bold')
ax.text(phase_7['t'].mean(), ax.get_ylim()[1] * 0.8, f'MAE: {error_phase7:.3f}', ha='center', fontsize=10, fontweight='bold')

# --- Panel 3: Outlier fraction and ESS ---
ax = axes[2]
ax.fill_between(window['t'], 0, window['outlier_frac'], alpha=0.5, color='orange', label='Outlier Frac')
ax.set_ylabel('Outlier Fraction', color='orange')
ax.axvspan(phase_6['t'].min(), phase_6['t'].max(), alpha=0.2, color='orange')
ax.axvspan(phase_7['t'].min(), phase_7['t'].max(), alpha=0.2, color='red')

ax2 = ax.twinx()
ax2.plot(window['t'], window['ess'], 'green', linewidth=1, label='ESS')
ax2.axhline(50, color='red', linestyle='--', alpha=0.5)
ax2.set_ylabel('ESS', color='green')

ax.set_xlabel('Time (tick)')

plt.tight_layout()
plt.savefig('critical_test.png', dpi=150, bbox_inches='tight')
plt.show()

# Print verdict
print("\n" + "="*60)
print("VERDICT:")
print("="*60)
print(f"Phase 6 (Outlier Burst, h should stay at -4.0):")
print(f"  - Mean h_est: {phase_6['h_est'].mean():.3f} (should be ~-4.0)")
print(f"  - h_est range: [{phase_6['h_est'].min():.3f}, {phase_6['h_est'].max():.3f}]")
print(f"  - MAE: {error_phase6:.3f}")
print(f"")
print(f"Phase 7 (Real Crash, h should jump to -2.0):")
print(f"  - Mean h_est: {phase_7['h_est'].mean():.3f} (should be ~-2.0)")
print(f"  - MAE: {error_phase7:.3f}")
print(f"")
if error_phase6 < 0.3 and phase_6['h_est'].mean() < -3.5:
    print(" GOOD: Filter did NOT overreact to outlier burst")
else:
    print(" BAD: Filter overreacted to outlier burst")
    
if error_phase7 < 0.5:
    print(" GOOD: Filter adapted to real crash")
else:
    print(" BAD: Filter was slow to adapt to real crash")
```

## 5. Summary Statistics

```python
print("\n" + "="*70)
print("OVERALL RBPF PERFORMANCE SUMMARY")
print("="*70)

error = df['h_est'] - df['h_true']

print(f"\nVolatility Tracking:")
print(f"  Overall RMSE:          {np.sqrt((error**2).mean()):.4f}")
print(f"  Overall Bias:          {error.mean():.4f}")
print(f"  95% CI Coverage:       {((df['h_true'] >= df['h_est'] - 2*df['h_std']) & (df['h_true'] <= df['h_est'] + 2*df['h_std'])).mean()*100:.1f}%")

print(f"\nRegime Detection:")
print(f"  Overall Accuracy:      {(df['regime_true'] == df['regime_est']).mean()*100:.1f}%")

print(f"\nFilter Stability:")
print(f"  Mean ESS:              {df['ess'].mean():.1f}")
print(f"  Min ESS:               {df['ess'].min():.1f}")
print(f"  ESS Collapses (<50):   {(df['ess'] < 50).sum()}")

print(f"\nOutlier Handling:")
print(f"  Total Outliers:        {df['is_outlier'].sum()}")
print(f"  Mean Outlier Frac:     {df['outlier_frac'].mean():.3f}")
print(f"  Max Outlier Frac:      {df['outlier_frac'].max():.3f}")

print("\n" + "="*70)
```
